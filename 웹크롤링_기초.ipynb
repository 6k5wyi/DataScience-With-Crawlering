## 멜론에서 노래 순위 정보 크롤링

위쪽에서 colab 또는 local에서 실행시켜야 할 부분을 체크하고(실행하고) 계속 진행

```python
url = 'http://www.melon.com/chart/index.htm'
wd.get(url)
```

```python
from bs4 import BeautifulSoup
html = wd.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
songs = soup.select('table > tbody > tr')
print(len(songs))
print(songs[0])
```

```python
# 1위곡 뽑아내기
song = songs[0]
```

```python
# 1위곡 제목 뽑아내기
title = song.select('a')
print(title)
print()
print('1위곡 의 갯수', len(title),'개')
```

```python
# 윗부분에서 1 보다 큰 수가 나왔으니 잘못 뽑았음..
# 그래서 윗경로(span) 을 추가하여 더 세분화 하여 뽑아보기
title = song.select('span > a')
print(title)
len(title)
```

```python
# 윗부분에서 1 보다 큰 수가 나왔으니 잘못 뽑았음..
# 그래서 윗경로(div.ellipsis.rank01) 을 추가하여 더 세분화 하여 뽑아보기

title = song.select('div.ellipsis.rank01 > span > a')
print(title)
len(title)
```

```python
# 윗부분에서 1이 나왔으니, 타이틀을 뽑기

title = song.select('div.ellipsis.rank01 > span > a')[0]
title.text
```

```python
# 가수이름을 같은 방식으로 뽑음.
singer = song.select('div.ellipsis.rank02 > span > a')
# 가수가 1명이 아닐때 모두 뽑아보기
p =''
for i in singer:
    p = p + i.text + ', '
singer = p[:-2]
print(singer)
```

```python
# 1위곡 부터 100위곡 까지 반복문을 통해 뽑아보기

for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > span > a')
    for i in singer:
        p = p + i.text + ', '
    singer = p[:-2]
    p = ''
    print(title, singer, sep=' | ') ## |을 구분자로 사용하여 프린트 하기
```

```python
## 1위곡 제목만 뽑아서 리스트에 넣기
titlelist = []
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    titlelist.append(title)

print(titlelist) 
```

```python
## 1위곡 제목과 가수를 뽑아서 리스트에 넣기 (구분은 | 으로 하기)
songlist = []
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > span > a')
    for i in singer:
        p = p + i.text + ', '
    singer = p[:-2]    
    p = ''
    songlist.append(title + ' | ' + singer)

print(songlist)
```
