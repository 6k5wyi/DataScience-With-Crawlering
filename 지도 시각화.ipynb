## ì¹´ì¹´ì˜¤ ë¡œì»¬ APIë¥¼ í™œìš©í•œ ì¥ì†Œ ê²€ìƒ‰ í•¨ìˆ˜

```python
def find_place(searching):
    # ì ‘ì† URL
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
    # í•´ë” ì…ë ¥
    headers = {
        "Authorization": "KakaoAK 1cf4f3d8027542a5cc811ea6aa45a554"
    }
    #API ìš”ì²­ ë° ì •ë³´ ë°›ê¸°
    places = requests.get(url, headers = headers).json()['documents']
    # í•„ìš”í•œ ì •ë³´ ì„ íƒ
    places = places[0]
    name = places['place_name']
    x = places['x']
    y = places['y']
    data = [name, x, y, searching]
    
    return data
```

# foliumì„ ì´ìš©í•œ ì§€ë„ ì‹œê°í™”

```python
location_data = pd.read_excel('./3_location_inform.xlsx')
location_data.info()
```

```python
import folium
Mt_Hanla = [33.362500,126.533694] # í•œë¼ì‚°ì„ ì„¼í„°ë¡œ ì¡ê¸°
map_jeju = folium.Map(location=Mt_Hanla, zoom_start = 10)

for i in range(len(location_data)):
    name = location_data['name_official'][i]
    count = location_data['place'][i]
    size = int(count)*2
    long = float(location_data['ìœ„ë„'][i])
    lat = float(location_data['ê²½ë„'][i])
    folium.CircleMarker((long,lat),radius = size , color = 'red', popup = name).add_to(map_jeju)
    
map_jeju
```

```python
# ì§€ë„ ì €ì¥
map_jeju.save('./3_jeju.html')
```

## ê·¸ë£¹ìœ¼ë¡œ í‘œì‹œí•˜ê¸°

```python
from folium.plugins import MarkerCluster

locations = []
name = []

for i in range(len(location_data)):
    data = location_data.iloc[i] # í–‰ í•˜ë‚˜ì”© ë½‘ì•„ì„œ dataì— ì €ì¥
    locations.append((float(data['ìœ„ë„']),float(data['ê²½ë„']))) # ìœ„ë„ ê²½ë„ ìˆœìœ¼ë¡œ
    name.append(data['name_official'])
    
Mt_Hanla = [33.362500,126.533694] # í•œë¼ì‚°ì„ ì„¼í„°ë¡œ ì¡ê¸°
map_jeju2 = folium.Map(location=Mt_Hanla, zoom_start = 10)

marker_cluster = MarkerCluster(
locations = locations, popups=name, name = 'Jeju', overlay=True, control = True)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)

map_jeju2
    
```

```python
## htmlë¡œ ì €ì¥
map_jeju2.save('./3_jeju_cluster.html')
```
